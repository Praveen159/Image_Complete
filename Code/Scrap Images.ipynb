```python
def generate_power(number):
    """
    Examples of use:

    >>> raise_two = generate_power(2)
    >>> raise_three = generate_power(3)
    >>> print(raise_two(7))
    128
    >>> print(raise_three(5))
    243
    """

    # Define the inner function ...
    def nth_power(power):
        return number ** power
    # ... that is returned by the factory function.

    return nth_power

a=generate_power(4)
a(2)
```

```python
img_dir="C:\\Users\\hi\\POC\\CarImages"
url="http://www.hdnicewallpapers.com/Cars-Wallpapers/7"
import requests

def car_imagess(urll,countt):
    driver=webdriver.Chrome(chromedriver)
    driver.get(urll)
    WebDriverWait(driver,5)
    
    def images_scrap(count):
        page=BeautifulSoup(driver.page_source,"html5lib")
        images1=page.findAll("img",{'src':re.compile(r'\S+.jpg')})
        urls=[img['src'] for img in images1]
        for url in urls:
            url_real="http://www.hdnicewallpapers.com"+url
            count=count+1
            try:
                with open(img_dir+str(count)+".jpg","wb") as output2:
                    response=requests.get(url_real)
                    output2.write(response.content)
            except:
                print("url not proper")
        return count   
    a=images_scrap(countt)            
    for i in range(1,15):
        driver.find_element_by_link_text("Next").click()
        b=images_scrap(a)
        a=b
        
        
            
            
    driver.close()
    return a
            
        
    
    
    
```
